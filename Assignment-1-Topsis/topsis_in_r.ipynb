```r
install.packages("topsis")
library(topsis)
```

```r
mydata <- read.csv('test_data.csv')
```

```r
criteria <- as.matrix(mydata[, -1])
```

```r
weights <- c(1, 1, 1, 1)
impacts <- c("+", "+", "-", "+")
```

```r
result <- topsis(criteria, weights, impacts)
```

```r
output <- data.frame(mydata,
                     TopsisScore = round(result$score, 3),
                     Rank = result$rank)
```

```r
library(readr)
write_csv(output, "results.csv")
print(output)
```
